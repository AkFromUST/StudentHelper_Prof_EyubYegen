```python
import pandas as pd

agencies_csv = "agencies_list/agencies.csv"
jurisdictions_csv = "agencies_list/jurisdictions.csv"

agencies_df = pd.read_csv(agencies_csv)
jurisdictions_df = pd.read_csv(jurisdictions_csv)

# print(agencies_df.columns)
# print(jurisdictions_df.columns)

jurisdictions_df["parent_abbrev"] = ""
```

Confirming that there are 55 abbrevs

```python
abbrevs = {}

for row in jurisdictions_df.to_dict(orient="records"):
    abbrev = row["abbrev"]
    parent = row["parent"]
    id = int(row["id"])

    if pd.notna(abbrev):
        abbrevs[id] = abbrev

for row in jurisdictions_df.to_dict(orient="records"):
    parent = row["parent"]

    if parent in abbrevs:
        #adding a new column to the df
        jurisdictions_df.loc[row["id"], "parent_abbrev"] = abbrevs[parent]

jurisdictions_df.to_csv("agencies_list/jurisdictions_with_abbrev.csv", index=False)
```
